Espectograma de 'emg_data10.csv'

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt, iirnotch, spectrogram

# Função para filtro passa-baixa
def filtro_passa_baixa(sinal, cutoff, fs, ordem=4):
    b, a = butter(ordem, cutoff / (0.5 * fs), btype='low')
    return filtfilt(b, a, sinal)

# Função para filtro notch (60 Hz)
def filtro_notch(sinal, freq_notch, fs, q=30):
    b, a = iirnotch(freq_notch, q, fs)
    return filtfilt(b, a, sinal)

# Carregar dados
dados = pd.read_csv('emg_data10.csv')

# Extrair dados
tempo = dados['Tempo(s)'].values
sinal_emg = dados['EMG_Value'].values

# Parâmetros
taxa_amostragem = 1000  # Exemplo: 1000 Hz

# Aplicar filtros
sinal_filtrado = filtro_passa_baixa(sinal_emg, cutoff=20, fs=taxa_amostragem)
sinal_filtrado = filtro_notch(sinal_filtrado, freq_notch=60, fs=taxa_amostragem)

# Gerar espectrograma
plt.figure(figsize=(10, 6))
f, t, Sxx = spectrogram(sinal_filtrado, fs=taxa_amostragem, nperseg=256, noverlap=128, scaling='density')
plt.pcolormesh(t, f, 10 * np.log10(Sxx), shading='gouraud')
plt.colorbar(label='Potência (dB/Hz)')
plt.title('Espectrograma do Sinal EMG Filtrado')
plt.ylabel('Frequência (Hz)')
plt.xlabel('Tempo (s)')
plt.ylim(0, 150)  # Frequências típicas do sinal EMG
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import csv
from scipy.signal import butter, filtfilt

DATA_FILENAME = "emg_data.csv"

# Função para filtro passa-alta
def high_pass_filter(data, cutoff=20, sample_rate=1000, order=4):
    nyquist = 0.5 * sample_rate
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='high', analog=False)
    return filtfilt(b, a, data)

# Carregar dados do arquivo CSV
def load_data(filename):
    times = []
    emg_values = []
    with open(filename, 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Pular cabeçalho
        for row in reader:
            times.append(float(row[0]))
            emg_values.append(int(row[1]))
    return np.array(times), np.array(emg_values)

# Analisar potência por faixa de frequência
def analyze_frequency_bands(emg_values, sample_rate=1000):
    n = len(emg_values)
    emg_values = emg_values - np.mean(emg_values)  # Remover componente DC
    emg_values = high_pass_filter(emg_values, cutoff=20)  # Filtro passa-alta

    fft_values = np.fft.fft(emg_values)
    freq = np.fft.fftfreq(n, d=1/sample_rate)

    # Apenas metade do espectro é necessária
    half_n = n // 2
    fft_values = np.abs(fft_values[:half_n])
    freq = freq[:half_n]

    # Cálculo da potência por faixa
    def power_in_band(f_min, f_max):
        mask = (freq >= f_min) & (freq < f_max)
        return np.sum(fft_values[mask]**2)

    power_total = np.sum(fft_values**2)
    bands = {
        "20-60 Hz": power_in_band(20, 60),
        "60-150 Hz": power_in_band(60, 150),
        "150-450 Hz": power_in_band(150, 450)
    }

    # Exibir resultados
    for band, power in bands.items():
        percentage = (power / power_total) * 100
        print(f"Faixa {band}: {percentage:.2f}% da potência total")

    # Gráfico do espectro
    plt.figure(figsize=(10, 6))
    plt.plot(freq, fft_values)
    plt.axvspan(20, 60, color='green', alpha=0.3, label='20-60 Hz (Inicial)')
    plt.axvspan(60, 150, color='blue', alpha=0.3, label='60-150 Hz (Principal)')
    plt.axvspan(150, 450, color='purple', alpha=0.3, label='150-450 Hz (Alta frequência)')
    plt.title("Espectro de Frequências com Faixas de Interesse")
    plt.xlabel("Frequência (Hz)")
    plt.ylabel("Magnitude")
    plt.legend()
    plt.grid(True)
    plt.show()

# Executar análise
try:
    times, emg_values = load_data(DATA_FILENAME)
    analyze_frequency_bands(emg_values)
except FileNotFoundError:
    print(f"Erro: Arquivo '{DATA_FILENAME}' não encontrado.")
except Exception as e:
    print(f"Ocorreu um erro: {e}")
```

Código usado para coletar os dados que possuem o noem emg_datax.csv


```python
from machine import ADC, Pin, Timer
import time
from time import sleep

# Configuração
adc_pin = 26
adc = ADC(Pin(adc_pin))
DATA_FILENAME = "emg_data.csv"
RECORDING_DURATION = 30  # segundos

# Buffer otimizado
BUFFER_SIZE = 5000  # Coleta mais eficiente para RAM e escrita no arquivo
emg_buffer = []
start_time = time.ticks_us()

messages = [
    (0, 'Manter o braço parado com a mão aberta'),
    (5, 'Feche a mão'),
    (10, 'Abra a mão'),
    (15, 'Feche a mão'),
    (20, 'Abra a mão'),
    (25, 'Feche a mão'),
]

last_message_index = -1

# Frequência de coleta aumentada
SAMPLE_RATE = 1000  # 1000 Hz (1 ms entre amostras)

def collect_data(timer):
    global emg_buffer, start_time, last_message_index

    emg_value = adc.read_u16()
    elapsed_time = time.ticks_diff(time.ticks_us(), start_time) / 1_000_000
    emg_buffer.append((elapsed_time, emg_value))

    for i, (msg_time, msg) in enumerate(messages):
        if elapsed_time >= msg_time and last_message_index < i:
            print(f'{elapsed_time:.1f}s: {msg}')
            last_message_index = i
            break

    # Salva quando o buffer atingir o tamanho máximo
    if len(emg_buffer) >= BUFFER_SIZE:
        save_buffer()
        emg_buffer = []  # Limpa o buffer

def save_buffer():
    with open(DATA_FILENAME, "a") as file:
        for time_val, emg_val in emg_buffer:
            file.write(f"{time_val:.6f},{emg_val}\n")

# Inicializa arquivo
with open(DATA_FILENAME, "w") as file:
    file.write("Tempo(s),EMG_Value\n")

# Configura Timer
timer = Timer()
timer.init(freq=SAMPLE_RATE, mode=Timer.PERIODIC, callback=collect_data)

try:
    print(f"Gravando por {RECORDING_DURATION} segundos...")
    sleep(RECORDING_DURATION)
    timer.deinit()  # Para o timer
    save_buffer()   # Salva dados restantes
    print("\nGravação finalizada!")
except KeyboardInterrupt:
    timer.deinit()
    save_buffer()
    print("\nGravação interrompida pelo usuário.")
```
